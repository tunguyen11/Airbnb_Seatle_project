Before doing the imputation for the categorical variable, I need to do the following: 
* Regroup the property_type column. The property_type column has 16 different value, but many values has very few samples. So I map multiple value of to the one group so that we don't have two many columns when we do One-Hot-Encode and we still have a good amount of information

#### Step 1.2.5  Featuring amenities column: 
Amenities column contains string of the amenities available in each listing, each separated from others by comma. I will convert this string type columns to a matrix of all the amenities available cross all the listings and assign binary value for each column of amenity

```python
# get the set of all amenities 
def replace_multiple(x):
    for char in ['{', '}',' ', '"']:
        x = x.replace(char, '')
    return x 
listing_data['amenities_cleaned'] = listing_data['amenities'].apply(lambda x: replace_multiple(x))

all_amenities = []
for i in listing_data['amenities_cleaned']:
    i_list = i.split(",")
    all_amenities.extend(i_list)
    amenities_set = set(all_amenities)
print(amenities_set)
    
```

```python
# create columns, that each one is an amenity in the amenities set: 
def amenity_present(x, amen):
    if amen in x:
        return 1
    else:
        return 0 

for amen in amenities_set:
    listing_data[amen] = listing_data['amenities'].apply(lambda x: amenity_present(x, amen))
    
```

### Step 1.3:  Choose the best model to predict listing price.

```python
# Look at the price data distribution 

sns.distplot(listing_data['price'])
```

The distribution of price data indicates that the price distribution is right-skewed. 

Therefore, I don't used the Linear Regression model for prediction. 

Rather, I prefer using other non-parametric methods such as Random Forest, K Neareast Neighboor 

```python
# create a function for evaluate a model with prediction time 
from sklearn.metrics import mean_squared_error, r2_score
from time import time

def train_predict(learner, X_train, y_train, X_test, y_test): 
    """
    inputs:
       - learner: the learning algorithm to be trained and predicted on
       - X_train: features training set
       - y_train: review_scores_rating training set
       - X_test: features testing set
       - y_test: review_scores_rating testing set
    """
    results = {}
    
    #Fit the learner to the training data and get training time
    start = time() 
    learner = learner.fit(X_train, y_train)
    end = time() 
    results['train_time'] = end-start
    
    # Get predictions on the test set(X_test), then get predictions on training samples
    start = time() 
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train)
    end = time() 
    
    # Calculate the total prediction time
    results['pred_time'] = end-start
    
    #Compute score on the training samples
    results['mse_train'] = mean_squared_error(y_train, predictions_train)
    results['r2_train'] = r2_score(y_train, predictions_train)
    
    #Compute accuracy on test set
    results['mse_test'] = mean_squared_error(y_test,predictions_test)
    results['r2_test'] = r2_score(y_test,predictions_test)
    
    print("MSE_train: %.4f" % results['mse_train'])
    print("MSE_test: %.4f" % results['mse_test'])
    print("Training score:%.4f" % learner.score(X_train,y_train))
    print("Test score:%.4f" % learner.score(X_test,y_test))
    return results
```

```python
#Initialize the model
from sklearn.ensemble import RandomForestRegressor 
from sklearn.neighbors import KNeighborsRegressor
clf1 = KNeighborsRegressor(random_state = 42)
clf2 = RandomForestRegressor(random_state = 42)

# Collect results on the learners
results = {}
for clf in [clf1, clf2]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    results[clf_name] = train_predict(clf,  X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
print(results['KNeighborsRegressor'])
print(results['RandomForestRegressor'])
```

From the result above, I prefer to use Random Forest model for predicting the price. 
In the next part, I will use GridSearchCV to find the best hyperparameter for the model.  

```python
# Implementation: Model Tuning for Random Forest Regressor 
# we don't need to use scale data since Random Forest learner doesn't require the data to be normalized 
from sklearn.model_selection import GridSearchCV 
clf = RandomForestRegressor(random_state = 42 )

# TODO: Create the parameters list you wish to tune, using a dictionary if needed.

parameters = {'max_depth': [5, 10, 15], 'min_samples_leaf': [4, 6,8, 10],  'n_estimators':[200, 300] ,'max_leaf_nodes' :[None, 2, 5, 10] }

grid = GridSearchCV(clf, param_grid = parameters, cv = 5, scoring = 'mean_squared_error' )
grid.fit(X_train, y_train)
grid.best_estimator_
grid.best_params_
grid.best_score_ 

# The optimized and model
best_clf = grid.best_estimator_

# Make predictions using the optimized and model
best_predictions = best_clf.predict(X_test)

# Make predictions using the unoptimized and model
predictions = clf.fit(X_train, y_train).predict(X_test )

# Report the before-and-afterscores
print("Unoptimized model\n------")
print("MSE on testing data: {:.4f}".format(mean_squared_error(y_test, predictions)))
print("R2 on testing data: {:.4f}".format(r2_score(y_test, predictions)))
print("\nOptimized Model\n------")
print("Final MSE on the testing data: {:.4f}".format(mean_squared_error(y_test, best_predictions)))
print("Final R2 on the testing data: {:.4f}".format(r2_score(y_test, best_predictions)))
```

```python
print(grid.best_params_)
print(best_clf)
```

###  Step 1.4 : Conclusion:  
From the result above I found the best model to predict price with optimized hyperparameter.

## Question 2:  What quantitative features has the most impact on the price?

###  Step 2.1:  Using the RandomForest Regressor to find the features importance 
The RandomForest Regressor learner has the feature_importance attribute. This attribute is a function that ranks the importance of each feature when making predictions based on the chosen algorithm.

```python
# TODO: Import a supervised learning model that has 'feature_importances_'
#Random Forest Classifier model has the feature_importances attribute 

# TODO: Train the supervised model on the training set using .fit(X_train, y_train)
model = RandomForestRegressor(random_state=42)

model.fit(X_train, y_train)

# TODO: Extract the feature importances using .feature_importances_ 
importances = model.feature_importances_
importances
```

###  Step 2.2:  Plot the top five important features 

```python
# Plot
# Create a function for feature plot 
def feature_plot(importances, X_train, y_train):
    
    # Display the most important features
    indices = np.argsort(importances)[::-1]
    columns = X_train.columns.values[indices[:5]]
    values = importances[indices][:5]

    # Creat the plot
    fig = plt.figure(figsize = (9,5))
    plt.title("Normalized Weights for First five Most Predictive Features", fontsize = 16)
    plt.bar(np.arange(5), values, width = 0.6, align="center", color = '#00A000', \
          label = "Feature Weight")
    plt.bar(np.arange(5) - 0.3, np.cumsum(values), width = 0.2, align = "center", color = '#00A0A0', \
          label = "Cumulative Feature Weight")
    plt.xticks(np.arange(5), columns)
    plt.xlim((-0.5, 4.5))
    plt.ylabel("Weight", fontsize = 12)
    plt.xlabel("Feature", fontsize = 12)
    
    plt.legend(loc = 'upper center')
    plt.tight_layout()
    plt.show()  
```

```python
feature_plot(importances, X_train, y_train)
```

###  Step 2.3 : Conclusion
From the graph above, we can see that, the most 5 important features in predicting the price is :
* number of bedrooms
* accomodates
* number of bathrooms

## Question 3 :  Can we use a machine learning to predict a review sentiment score from the review text? 


### Step 3.1 Use the pretrained model and the text review to predict the sentiment score 
SentimentIntensityAnalyzer can produce the sentiment score range from -1 to 1 that the higher score means the more positve the text review is and vice versa. 

```python
reviews.isnull().sum(axis = 0)
```

```python
reviews = reviews.dropna()
reviews.shape
```

```python
# Obtain the sentiment score value from the pretrained model 

from nltk.sentiment.vader import SentimentIntensityAnalyzer
reviews['scores'] = reviews['comments'].apply(lambda x: SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
```

```python
#group the review by listing id to find the average sentiment score of all review text pertaining to a listing id 
textreview_summary = reviews.groupby('listing_id')['scores'].mean()
print(type(textreview_summary))
```

```python
textreview_summary = textreview_summary.reset_index()
textreview_summary.head()
```

### Step 3.2 Combine with the review_score_rating to see how the predicted sentiment score in relation with the review_scores_rating move

```python
listing_review = listing_data[['id', 'review_scores_rating']]
listing_review['review_scores_rating'].value_counts()
```

```python
# merge the predict sentiment score with review_scores_value by the listing id 

review_combine = listing_review.merge(textreview_summary, left_on= 'id', right_on = 'listing_id')
review_combine = review_combine.drop('id', axis =1)
review_combine = review_combine.rename({'scores': 'sentiment_scores'})
review_combine.head()
```

```python
# check the correlation of review_score_rating and sentiment score
review_combine['review_scores_rating'].corr(review_combine['scores'])
```

```python
# Plot the review_score_rating and sentiment_score that predicted from pretrained model on the review text 

i = np.arange(review_combine.shape[0])
fig, ax1 = plt.subplots(figsize = (20, 10))

color = 'tab:red'
ax1.set_xlabel('listing')
ax1.set_ylabel('review_score_rating', color=color)
ax1.plot(i, review_combine['review_scores_rating'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sentiment_scores', color=color)  # we already handled the x-label with ax1
ax2.plot(i, review_combine['scores'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

### Step 3.3 Conclusion 

From the correlation result and the plot, we can see the the predicted sentiment score does not line up well with the review_score_rating
